```python
from fast_dash import FastDash
from fast_dash.Components import Image, UploadImage, Text
from keras.utils import load_img, img_to_array
from keras.utils import load_img
from keras.applications.inception_resnet_v2 import preprocess_input
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder
import os
import tensorflow as tf
import base64
import numpy as np

def preprocess_and_decode(img_str, new_shape=[128,128]):
    image_type, img_str = img_str.split(',')
    img = base64.b64decode(img_str)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, new_shape, method=tf.image.ResizeMethod.BILINEAR)
    img = img/255.
    # if you need to squeeze your input range to [0,1] or [-1,1] do it here
    return img
def transform_image(image):
    img = img_to_array(image)  # Convertir en tableau numpy
    img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))  # Créer la collection d'images (un seul échantillon)
    #img = preprocess_input(img)  # Prétraiter l'image comme le veut VGG-16
    return img

def predict(image) : 
    os.chdir('..')
    directory = "model_to_load" 
    os.chdir(directory)
    model = keras.models.load_model('Model_trained_2')
    image = model.predict(image)
    return image

def untransform_image(prediction):
    directory = "Encoder" 
    os.chdir(directory)
    encoder = LabelEncoder()
    encoder.classes_ = np.load('classes.npy')
    
    dog_class = encoder.inverse_transform([np.argmax(prediction)])
    return dog_class

def image_to_image(image):
    image = preprocess_and_decode(image)
    image = transform_image(image)
    prediction = predict(image)
    dog_class = untransform_image(prediction)
    return dog_class

# 2. Define components and initialize app
app = FastDash(callback_fn=image_to_image, 
               inputs=UploadImage, 
               outputs=Text,
               title='Dog Classification',
               theme='FLATLY')
# 3. Deploy!
app.run()
```
